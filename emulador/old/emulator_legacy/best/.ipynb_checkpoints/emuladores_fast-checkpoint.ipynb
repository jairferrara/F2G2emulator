```python
"""
emuladores_fast.py
==================
Emulador ultrarrápido de A, B, A', B' usando ONNX Runtime.

PRIMER USO — convierte los modelos .keras a .onnx (solo una vez):
    python emuladores_fast.py --convertir

USO NORMAL desde tu notebook:
    from emuladores_fast import predict, predict_batch

    A, B, Ap, Bp = predict(0.3, 0.1, 0.05, -0.5, 0.28, -3.8)

    # Muchos puntos a la vez (aún más rápido por punto):
    X = np.array([[...], [...]])   # shape (N, 6)
    res = predict_batch(X)         # dict con 'A','B','Ap','Bp'
"""

import os, warnings, subprocess
import numpy as np
import joblib

warnings.filterwarnings("ignore")
os.environ["TF_CPP_MIN_LOG_LEVEL"] = "3"

# ── Ajusta estas rutas a donde tengas tus archivos ──────────────────────────
DIR = os.path.dirname(os.path.abspath(__file__)) if "__file__" in dir() else os.getcwd()

PATHS = {
    "keras_AB":    os.path.join(DIR, "best_emulator_AB.keras"),
    "keras_Ap":    os.path.join(DIR, "best_emulator_AprimeBprime.keras"),
    "onnx_AB":     os.path.join(DIR, "model_AB.onnx"),
    "onnx_Ap":     os.path.join(DIR, "model_Ap.onnx"),
    "saved_AB":    os.path.join(DIR, "_savedmodel_AB"),
    "saved_Ap":    os.path.join(DIR, "_savedmodel_Ap"),
    "scaler_X_AB": os.path.join(DIR, "scaler_X_AB.pkl"),
    "scaler_Y_AB": os.path.join(DIR, "scaler_Y_AB.pkl"),
    "scaler_X_Ap": os.path.join(DIR, "scaler_X_AprimeBprime.pkl"),
    "scaler_Y_Ap": os.path.join(DIR, "scaler_Y_AprimeBprime.pkl"),
}
# ────────────────────────────────────────────────────────────────────────────


def convertir():
    """
    Convierte los modelos .keras → .onnx.
    Ejecutar solo una vez. Requiere tensorflow y tf2onnx instalados.
    """
    import tensorflow as tf
    import keras

    @keras.saving.register_keras_serializable()
    class CosineWarmupSchedule(tf.keras.optimizers.schedules.LearningRateSchedule):
        def __init__(self, initial_lr, total_epochs, warmup_epochs,
                     steps_per_epoch, min_lr=1e-7, **kwargs):
            super().__init__(**kwargs)
            self.initial_lr      = initial_lr
            self.total_epochs    = total_epochs
            self.warmup_epochs   = warmup_epochs
            self.steps_per_epoch = steps_per_epoch
            self.min_lr          = min_lr
        def __call__(self, step): return self.initial_lr
        def get_config(self):
            return {"initial_lr": self.initial_lr, "total_epochs": self.total_epochs,
                    "warmup_epochs": self.warmup_epochs,
                    "steps_per_epoch": self.steps_per_epoch, "min_lr": self.min_lr}

    custom = {"CosineWarmupSchedule": CosineWarmupSchedule}
    print("Cargando modelos Keras...")
    m_AB = tf.keras.models.load_model(PATHS["keras_AB"], custom_objects=custom)
    m_Ap = tf.keras.models.load_model(PATHS["keras_Ap"], custom_objects=custom)

    print("Guardando como SavedModel (paso intermedio)...")
    tf.saved_model.save(m_AB, PATHS["saved_AB"])
    tf.saved_model.save(m_Ap, PATHS["saved_Ap"])

    print("Convirtiendo a ONNX...")
    for key_saved, key_onnx, label in [
        ("saved_AB", "onnx_AB", "AB"),
        ("saved_Ap", "onnx_Ap", "A'B'"),
    ]:
        r = subprocess.run(
            ["python3", "-m", "tf2onnx.convert",
             "--saved-model", PATHS[key_saved],
             "--output",      PATHS[key_onnx],
             "--opset",       "13"],
            capture_output=True, text=True
        )
        if r.returncode != 0:
            raise RuntimeError(f"Error convirtiendo {label}:\n{r.stderr}")
        print(f"  ✓ model_{label}.onnx generado")

    print("\n¡Conversión completada! Ya puedes usar predict() sin necesitar TensorFlow.\n")


# ── Motor de inferencia ONNX ─────────────────────────────────────────────────
_cache = {}

def _load_onnx():
    """Carga sesiones ONNX y scalers (solo la primera vez)."""
    if _cache:
        return

    # Verifica que existan los .onnx
    for key in ("onnx_AB", "onnx_Ap"):
        if not os.path.exists(PATHS[key]):
            raise FileNotFoundError(
                f"No se encontró {PATHS[key]}.\n"
                "Ejecuta primero:  python emuladores_fast.py --convertir"
            )

    import onnxruntime as rt

    opts = rt.SessionOptions()
    opts.intra_op_num_threads        = 1          # óptimo para 1 punto
    opts.graph_optimization_level    = rt.GraphOptimizationLevel.ORT_ENABLE_ALL
    opts.execution_mode              = rt.ExecutionMode.ORT_SEQUENTIAL

    providers = ["CPUExecutionProvider"]

    _cache["sess_AB"] = rt.InferenceSession(PATHS["onnx_AB"], sess_options=opts, providers=providers)
    _cache["sess_Ap"] = rt.InferenceSession(PATHS["onnx_Ap"], sess_options=opts, providers=providers)
    _cache["iname_AB"] = _cache["sess_AB"].get_inputs()[0].name
    _cache["iname_Ap"] = _cache["sess_Ap"].get_inputs()[0].name

    _cache["scaler_X_AB"] = joblib.load(PATHS["scaler_X_AB"])
    _cache["scaler_Y_AB"] = joblib.load(PATHS["scaler_Y_AB"])
    _cache["scaler_X_Ap"] = joblib.load(PATHS["scaler_X_Ap"])
    _cache["scaler_Y_Ap"] = joblib.load(PATHS["scaler_Y_Ap"])

    # Warm-up (elimina la latencia de la primera llamada real)
    dummy = np.zeros((1, 6), dtype=np.float32)
    _cache["sess_AB"].run(None, {_cache["iname_AB"]: dummy})
    _cache["sess_Ap"].run(None, {_cache["iname_Ap"]: dummy})


def predict(x1, x2, x3, x4, x5, x6):
    """
    Evalúa un solo punto.

    Retorna
    -------
    A, B, Ap, Bp : floats
    """
    _load_onnx()
    X = np.array([[x1, x2, x3, x4, x5, x6]], dtype=np.float64)

    X_AB = _cache["scaler_X_AB"].transform(X).astype(np.float32)
    X_Ap = _cache["scaler_X_Ap"].transform(X).astype(np.float32)

    Y_AB = _cache["scaler_Y_AB"].inverse_transform(
               _cache["sess_AB"].run(None, {_cache["iname_AB"]: X_AB})[0])
    Y_Ap = _cache["scaler_Y_Ap"].inverse_transform(
               _cache["sess_Ap"].run(None, {_cache["iname_Ap"]: X_Ap})[0])

    return float(Y_AB[0, 0]), float(Y_AB[0, 1]), float(Y_Ap[0, 0]), float(Y_Ap[0, 1])


def predict_batch(X):
    """
    Evalúa N puntos en paralelo (mucho más eficiente que N llamadas a predict).

    Parámetros
    ----------
    X : array-like, shape (N, 6)

    Retorna
    -------
    dict {"A": array(N), "B": array(N), "Ap": array(N), "Bp": array(N)}
    """
    _load_onnx()
    X = np.atleast_2d(np.asarray(X, dtype=np.float64))

    X_AB = _cache["scaler_X_AB"].transform(X).astype(np.float32)
    X_Ap = _cache["scaler_X_Ap"].transform(X).astype(np.float32)

    Y_AB = _cache["scaler_Y_AB"].inverse_transform(
               _cache["sess_AB"].run(None, {_cache["iname_AB"]: X_AB})[0])
    Y_Ap = _cache["scaler_Y_Ap"].inverse_transform(
               _cache["sess_Ap"].run(None, {_cache["iname_Ap"]: X_Ap})[0])

    return {"A": Y_AB[:, 0], "B": Y_AB[:, 1], "Ap": Y_Ap[:, 0], "Bp": Y_Ap[:, 1]}


def benchmark(n=1000):
    """Mide el tiempo promedio por llamada."""
    import time
    _load_onnx()

    # Single point
    times = []
    for _ in range(n):
        t0 = time.perf_counter()
        predict(0.3, 0.1, 0.05, -0.5, 0.28, -3.8)
        times.append(time.perf_counter() - t0)
    t_single = np.mean(times) * 1e6
    print(f"predict() — 1 punto:      {t_single:.1f} µs  ({t_single/1000:.3f} ms)")

    # Batch
    X_batch = np.tile([0.3, 0.1, 0.05, -0.5, 0.28, -3.8], (100, 1))
    times = []
    for _ in range(n // 10):
        t0 = time.perf_counter()
        predict_batch(X_batch)
        times.append(time.perf_counter() - t0)
    t_batch = np.mean(times) * 1e6
    print(f"predict_batch() — 100 pts: {t_batch:.1f} µs  ({t_batch/100:.1f} µs/punto)")
```

```python
# ── CLI ───────────────────────────────────────────────────────────────────────
if __name__ == "__main__":
    import sys

    if "--convertir" in sys.argv:
        convertir()

    elif "--benchmark" in sys.argv:
        print("Ejecutando benchmark...\n")
        benchmark()

    else:
        # Demo rápida
        print("=== Demo ===")
        import time
        A, B, Ap, Bp = predict(0.3, 0.1, 0.05, -0.5, 0.28, -3.8)
        print(f"  A  = {A:.6f}")
        print(f"  A' = {Ap:.6f}")
        print(f"  B  = {B:.6f}")
        print(f"  B' = {Bp:.6f}")
        print("\nBenchmark:")
        benchmark(500)
```
